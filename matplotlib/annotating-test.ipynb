```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ax = plt.subplot()
t = np.arange(0.0, 2.0, 0.01)
s = np.cos(2 * np.pi * t)
line, = plt.plot(t, s, lw=2)

plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
             arrowprops=dict(facecolor='black', shrink=0.05),
             )

plt.ylim(-2, 2)
plt.show()
```
